```python
# importar módulos
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("T1_Housing.csv",
                     delimiter=',',
                     skip_header=1)
print(data)
```

```python
#X = Avg Area Income
X = data[:,1]

#In = in sample

In_sample = data[:,3]

X_in = X * In_sample

X_out = X - X_in
X_out = X_out[X_out>0]
X_in = X_in[X_in>0]
#Y = Precio de las casas
Y = data[:,2]

Y_in = Y * In_sample

Y_out = Y - Y_in

Y_in = Y_in[Y_in>0]

Y_out = Y_out[Y_out>0]
```

```python
# Implementación de función reg_lineal(X,Y,rho)
def reg_lineal(X,Y,rho):
    """Devuelve el vector de paámetros theta utilizando regularización ridge"""
    X1 = np.vstack([X,np.ones_like(X)]).T 
    I = np.eye(len(X1.T))
    A = np.linalg.inv(np.dot(X1.T,X1)+(I*rho))
    B = np.dot(X1.T,Y)
    return A.dot(B)  
```

```python
thetha = reg_lineal(X_in, Y_in, 5)
```

```python
thetha
```

```python
rho = np.linspace(0 , 10 , 100)
theta = np.zeros(0)
for r in rho:
    theta_r = reg_lineal(X_in, Y_in, r)
    theta = np.append(theta, theta_r)
```
